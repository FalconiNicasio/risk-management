## Summary

Data Science can play a huge role in Risk Management. This project shows how to use approximately 60 million rows of historical transactional data from multiple data sources in order to predict the default risk of a home credit loan applicant. Engineering of over 600 features, gradient boosted tree model XGBoost, and Bayesian Hyperparameter Optimization were used which resulted in an Area Under the ROC Curve (AUC) score of 0.795. This is an increase of 0.051 over traditional Logistic Regression.

## Discussion

This project was done in 4 weeks and with the complexity and amount of data this problem presented, XGBoost was chosen to be used because of how robust it is to things like missing values, outliers, and unbalanced datasets. This provides a relatively "clean" start to complex problems that yields decent results especially when given a short amount of time. As a decision tree based model, this also gives interpretability to the features being used having the ability to measure the feature importance the model uses to split the data while building its trees.

It should also be noted that GPU training was utilized for this project which moved the bottle neck of the project from modeling and hyperparameter tuning (which could take up to 10x longer) to the ETL process (extract, transform, and load).

## Results and Conclusion

Overall, this project shows promise with the approach taken with feature engineering and choice of the model. It showed significant increases in AUC from 0.744 to 0.795 when adding more features (631) from other data sources and also using a more complex model over logistic regression. What this translates to is the user of this model has more accurate predictions given over any threshold used for the probability of an client to default on their home loan. For example if the threshold is held at 0.5, then the XGBoost would outperform Logistic Regression in its predictions and would give less false positives which would lead to better decisions when analyzing the risk of each applicant. Additionally, the model returns feature importance which could be used in the future approval process of each loan. The following will list the top 5 features engineered in this project.

### Top Features Engineered

`EXT_SOURCES_MEAN` - The mean of the three different credit scores combined.
 
`PREV_REFUSED_APP_CREDIT_RATIO_MIN` - Previous applications from the client that were refused and the minimum amount requested for those loans.

`PREV_APPROVED_APP_CREDIT_DIFF_MAX` - Previous applications from the client that were approved and the maximum of the difference of amount requested to the amount credit given.

`PREV_LATE_DAYS_DECISION_MEAN` - Previous applications that had late payments in their installments and the mean of the number of days it took to approve the loan.

`LAST_LOAN_PAID_OVER_AMOUNT_MEAN` - The mean of combined installment payments that the client paid over the minimum amount.

## Recommendations

This dataset is very rich in content and has limitless potential for feature engineering. More features could be explored and use of time series techniques could possibly be explored for monthly balance data like credit card balance and installment payments to find trends and predictions for missing current data. Clustering techniques could also be explored to try and group applicants into meaningful groups as a feature which could show improvements in the model. Lastly, while XGBoost is capable of splitting data with missing values, imputation techniques like KNN Imputer could be explored to see if imputing the data would show improvments in AUC.

## Helper functions

```python
def train_(model, X_train, y_train, X_test, y_test):
    """function to train XGBoost model
    Args:
        model (XGBoost.model): gradient boosted tree model
        X_train (pd.DataFrame): train data
        y_train (pd.DataFrame): train target
        X_test (pd.DataFrame): test/validation data
        y_test (pd.DataFrame): test/validation target
    """
    model.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_test, y_test)], 
          eval_metric=['logloss', 'auc'], early_stopping_rounds=5, verbose=False)
    y_pred_proba = model.predict_proba(X_test)[:,1]
    y_pred = model.predict(X_test)
    score = roc_auc_score(y_test, y_pred_proba)
    #acc = (y_pred==y_test).sum() / len(y_test)
    return score
```

```python
def do_exp(df, feat, params=None):
    """function to run experiment with XGBoost
    Args:
          df (pd.DataFrame): train|test application
          feats (pd.DataFrame): features engineered to add to train|test application
          params (dict): **kwargs for XGBoostClassifier

    Returns:
          cols (dict): columns and scores
    """
    #train = preprocess(df, col)
    X, y = df.drop(['TARGET'], axis=1), df.TARGET
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=27)

    if params is not None:
        model = xgb.XGBClassifier(tree_method='gpu_hist', n_estimators=1000 ,seed=27, 
                              use_label_encoder=False, **params)
    else:
        model = xgb.XGBClassifier(tree_method='gpu_hist', n_estimators=1000 ,seed=27, 
                              use_label_encoder=False, learning_rate=0.1, scale_pos_weight=3.35)

    score = train_(model, X_train, y_train, X_test, y_test)
    best_score = score

    feat_cols = list(feat.columns)
    cols = dict()

    t = tqdm(feat_cols)
    for col in t:
        #col = feat_cols[i]
        t.set_description(f"Testing {col}..")
        t.refresh() # to show immediately the update
        sleep(0.01)

        X_train_feat = X_train.merge(feat[[col]], how='left', on='SK_ID_CURR')
        X_test_feat = X_test.merge(feat[[col]], how='left', on='SK_ID_CURR')

        new_score = train_(model, X_train_feat, y_train, X_test_feat, y_test)
        if new_score >= score:
            #print(colored(f"{col} is a good feature!", 'green'))
            cols[col] = new_score-score
        
        if new_score > best_score:
            best_score = new_score
    print(f"best score: {best_score}")

    return cols
```

```python
def get_group_stats(df, group_cols, counted, agg_name, col_name):
    """function to group data and get aggregate statistics of categorical data
    Args:
        df(pd.DataFrame): data
        group_cols(list): list of columns names to group by
        counted(list): list of columns to aggregate by
        agg_name(str): aggregation name for aggregate function
        col_name(str): name of the new column/feature engineered
    """
    gp = df[group_cols + [counted]].groupby(group_cols)[counted].agg(agg_name).reset_index().rename(
        columns={counted: col_name})
    df = df.merge(gp, on=group_cols, how='left').set_index('SK_ID_CURR')
    del gp
    gc.collect()
    return df

def group(df, prefix, aggregations, group_by='SK_ID_CURR'):
    """function to group data and get aggregate statistics of numerical data
    Args:
        df(pd.DataFrame): dataframe to aggregate
        prefix(string): column/feature prefix to add to column name
        aggregations(dict): mapped aggregations to features in a dictionary
        aggregate_by(str): column name to group by
    """
    agg_df = df.groupby(group_by).agg(aggregations)
    agg_df.columns = pd.Index(['{}{}_{}'.format(prefix, e[0], e[1].upper())
                               for e in agg_df.columns.tolist()])
    return agg_df.reset_index()
```

```python
def get_age_label(days_birth):
    """function to return the age group label (int)
    Args:
        days_birth(np.array): DAYS_BIRTH column from application.csv table
    """
    age_years = -days_birth / 365
    if age_years < 27: return 1
    elif age_years < 40: return 2
    elif age_years < 50: return 3
    elif age_years < 65: return 4
    elif age_years < 99: return 5
    else: return 0
```

## application Train|Test

This will also serve as the base features for experiments

```python
def get_base(path, test=False):
    """get base features from application_train|test files
    path(str): path to directory where data is in
    test(boolean): if True will read in test data instead of train
    """
    if test:
        df = pd.read_csv(os.path.join(path, 'application_test.csv.zip'), index_col='SK_ID_CURR')
    else:
        df = pd.read_csv(os.path.join(path, 'application_train.csv.zip'), index_col='SK_ID_CURR')

    #cleaning
    df.replace(365243, float('inf'), inplace=True)
    df[df.dtypes[df.dtypes=='O'].index] = df[df.dtypes[df.dtypes=='O'].index].fillna('XNA')

    #encode------------------------------------------------------------------#
    bin_cols = []
    oth_cols = []

    for col in list(df[df.dtypes[df.dtypes=='O'].index].columns):
        if len(df[col].value_counts()) == 2:
            bin_cols.append(col)
        else:
            oth_cols.append(col)

    lb = LabelBinarizer()
    for col in bin_cols:
        df[col] = lb.fit_transform(df[col])

    le = LabelEncoder()
    for col in oth_cols:
        df[col] = le.fit_transform(df[col])

    #df = pd.get_dummies(df)
    return df
```

```python
def get_app(path, test=False):
    """preprocesses application train|test data
    Args:
        path(str): path to directory where data is in
        test(boolean): if True will read in test data instead of train
    Returns:
        df(pd.DataFrame): dataframe of preproccessed application train|test
                          data with features engineered
    """
    df = get_base(path, test=test)
    #------------------------------------------------------------------------#
    df['AGE_RANGE'] = df['DAYS_BIRTH'].apply(lambda x: get_age_label(x))

    # Credit ratios
    df['CREDIT_TO_ANNUITY_RATIO'] = df['AMT_CREDIT'] / (df['AMT_ANNUITY'] + 1)
    df['CREDIT_TO_GOODS_RATIO'] = df['AMT_CREDIT'] / (df['AMT_GOODS_PRICE'] + 1)
    df['CREDIT_TO_INCOME_RATIO'] = df['AMT_CREDIT'] / df['AMT_INCOME_TOTAL']
    # Income ratios
    df['ANNUITY_TO_INCOME_RATIO'] = df['AMT_ANNUITY'] / df['AMT_INCOME_TOTAL']
    df['CREDIT_TO_INCOME_RATIO'] = df['AMT_CREDIT'] / df['AMT_INCOME_TOTAL']
    df['INCOME_TO_EMPLOYED_RATIO'] = df['AMT_INCOME_TOTAL'] / df['DAYS_EMPLOYED']
    df['INCOME_TO_BIRTH_RATIO'] = df['AMT_INCOME_TOTAL'] / df['DAYS_BIRTH']
    # Time ratios
    #df['EMPLOYED_TO_BIRTH_RATIO'] = df['DAYS_EMPLOYED'] / df['DAYS_BIRTH']
    #df['ID_TO_BIRTH_RATIO'] = df['DAYS_ID_PUBLISH'] / df['DAYS_BIRTH']
    df['CAR_TO_BIRTH_RATIO'] = df['OWN_CAR_AGE'] / df['DAYS_BIRTH']
    #df['CAR_TO_EMPLOYED_RATIO'] = df['OWN_CAR_AGE'] / df['DAYS_EMPLOYED']
    df['PHONE_TO_BIRTH_RATIO'] = df['DAYS_LAST_PHONE_CHANGE'] / df['DAYS_BIRTH']
    df['PHONE_TO_EMP_RATIO'] = df['DAYS_LAST_PHONE_CHANGE'] / df['DAYS_EMPLOYED']


    #------------------------------------------------------------------------#
    docs = [_f for _f in df.columns if 'FLAG_DOC' in _f]
    live = [_f for _f in df.columns if ('FLAG_' in _f) & ('FLAG_DOC' not in _f) & ('_FLAG_' not in _f)]

    inc_by_org = df[['AMT_INCOME_TOTAL', 'ORGANIZATION_TYPE']].groupby('ORGANIZATION_TYPE').median()['AMT_INCOME_TOTAL']

    df['DOCUMENT_COUNT'] = df[docs].sum(axis=1)
    #df['DOC_IND_KURT'] = df[docs].kurtosis(axis=1)
    df['LIVE_IND_SUM'] = df[live].sum(axis=1)
    df['INC_PER_CHLD'] = df['AMT_INCOME_TOTAL'] / (1 + df['CNT_CHILDREN'])
    df['INC_BY_ORG'] = df['ORGANIZATION_TYPE'].map(inc_by_org)
    #df['EMPLOY_TO_BIRTH_RATIO'] = df['DAYS_EMPLOYED'] / df['DAYS_BIRTH']
    df['EXT_SOURCES_PROD'] = df['EXT_SOURCE_1'] * df['EXT_SOURCE_2'] * df['EXT_SOURCE_3']
    for function_name in ['min', 'max', 'mean', 'nanmedian']:
        feature_name = 'EXT_SOURCES_{}'.format(function_name.upper())
        df[feature_name] = eval('np.{}'.format(function_name))(
            df[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']], axis=1)
    df['EXT_SCORES_STD'] = df[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']].std(axis=1)
    df['EXT_SCORES_STD'] = df['EXT_SCORES_STD'].fillna(df['EXT_SCORES_STD'].mean())
    df['EXT_SOURCES_WEIGHTED'] = df.EXT_SOURCE_1 * 2 + df.EXT_SOURCE_2 * 1 + df.EXT_SOURCE_3 * 3

    # Groupby: Statistics for applications in the same group
    cols = ['ORGANIZATION_TYPE', 'NAME_EDUCATION_TYPE', 'OCCUPATION_TYPE', 'AGE_RANGE', 'CODE_GENDER']
    df = get_group_stats(df.reset_index(), cols, 'EXT_SOURCES_MEAN', 'median', 'GROUP_EXT_SOURCES_MEDIAN')
    df = get_group_stats(df.reset_index(), cols, 'EXT_SOURCES_MEAN', 'std', 'GROUP_EXT_SOURCES_STD')
    df = get_group_stats(df.reset_index(), cols, 'AMT_INCOME_TOTAL', 'mean', 'GROUP_INCOME_MEAN')
    df = get_group_stats(df.reset_index(), cols, 'AMT_INCOME_TOTAL', 'std', 'GROUP_INCOME_STD')
    #df = get_group_stats(df.reset_index(), cols, 'CREDIT_TO_ANNUITY_RATIO', 'mean', 'GROUP_CREDIT_TO_ANNUITY_MEAN')
    df = get_group_stats(df.reset_index(), cols, 'CREDIT_TO_ANNUITY_RATIO', 'std', 'GROUP_CREDIT_TO_ANNUITY_STD')
    df = get_group_stats(df.reset_index(), cols, 'AMT_CREDIT', 'mean', 'GROUP_CREDIT_MEAN')
    df = get_group_stats(df.reset_index(), cols, 'AMT_ANNUITY', 'mean', 'GROUP_ANNUITY_MEAN')
    df = get_group_stats(df.reset_index(), cols, 'AMT_ANNUITY', 'std', 'GROUP_ANNUITY_STD')
    
    #return df.drop(cols, axis=1)
    return df
```

## installments_payments

```python
def get_ip(path, nrows=None):
    """preprocesses installments_payments data
    Args:
        path(str): path to directory where data is in
        nrows(int): number of rows to read in (used for testing purposes)
    Returns:
        df(pd.DataFrame): dataframe of preproccessed installments_payments
                          data with features engineered
    """
    INSTALLMENTS_AGG = {
        'SK_ID_PREV': ['size', 'nunique'],
        'DAYS_ENTRY_PAYMENT': ['max', 'mean'],
        'AMT_INSTALMENT': ['min', 'max', 'mean', 'sum'],
        'AMT_PAYMENT': ['min', 'max', 'mean'],
        'DPD': ['max', 'mean', 'var', 'sum'],
        'UNDERPAID_PAYMENTS': ['sum'],
        'PAYMENT_DIFFERENCE': ['mean'],
        #'PAYMENT_RATIO': ['mean'],
        #'PAID_OVER': ['mean'],
        'DBD': ['var', 'sum'],
        'LATE_PAYMENT': ['mean', 'sum'],
        'INSTALMENT_PAYMENT_RATIO': ['mean'],
        'LATE_PAYMENT_RATIO': ['mean'],
        'SIGNIFICANT_LATE_PAYMENT': ['mean', 'sum'],
        #'DPD_7': ['mean'],
        #'DPD_15': ['mean'],
        }

    df = pd.read_csv(os.path.join(path, 'installments_payments.csv.zip'), nrows=nrows)

    # days past due
    df['DPD'] = (df['DAYS_INSTALMENT'] - df['DAYS_ENTRY_PAYMENT']).clip(lower=0)
    # underpaid
    df["UNDERPAID_PAYMENTS"] = (df.AMT_INSTALMENT > df.AMT_PAYMENT).astype('int8')

    df = get_group_stats(df, ['SK_ID_PREV', 'NUM_INSTALMENT_NUMBER'], 'AMT_PAYMENT', 'sum', 'AMT_PAYMENT_GROUPED')
    df['PAYMENT_DIFFERENCE'] = df['AMT_INSTALMENT'] - df['AMT_PAYMENT_GROUPED']
    df['PAYMENT_RATIO'] = df['AMT_INSTALMENT'] / df['AMT_PAYMENT_GROUPED']
    df['PAID_OVER_AMOUNT'] = df['AMT_PAYMENT'] - df['AMT_INSTALMENT']
    df['PAID_OVER'] = (df['PAID_OVER_AMOUNT'] > 0).astype('int8')
    #days before due
    df['DBD'] = (df['DAYS_INSTALMENT'] - df['DAYS_ENTRY_PAYMENT']).clip(lower=0)
    #flag late payment
    df['LATE_PAYMENT'] = (df['DPD'] > 0).astype('int8')
    # percentage of payments that were late
    df['INSTALMENT_PAYMENT_RATIO'] = df['AMT_PAYMENT'] / df['AMT_INSTALMENT']
    df['LATE_PAYMENT_RATIO'] = df.where(df.LATE_PAYMENT==1)['INSTALMENT_PAYMENT_RATIO'].fillna(0)
    # Flag late payments that have a significant amount
    df['SIGNIFICANT_LATE_PAYMENT'] = (df['LATE_PAYMENT_RATIO'] > 0.05).astype('int8')
    # flag k threshold late payments
    #df['DPD_7'] = df['DPD'].apply(lambda x: 1 if x>=7 else 0)
    #df['DPD_15'] = df['DPD'].apply(lambda x: 1 if x>= 15 else 0)

    # Aggregations by SK_ID_CURR
    pay_agg = group(df, 'INS_', INSTALLMENTS_AGG) #agg df

    INSTALLMENTS_36_AGG = {
        'SK_ID_PREV': ['size'],
        'DAYS_ENTRY_PAYMENT': ['max', 'mean'],
        'AMT_INSTALMENT': ['min', 'max', 'mean', 'sum'],
        'AMT_PAYMENT': ['min', 'max', 'sum'],
        'DPD': ['var'],
        'DBD': ['max', 'mean', 'var'],
        'PAYMENT_DIFFERENCE': ['mean'],
        #'PAYMENT_RATIO': ['mean'],
        'LATE_PAYMENT': ['mean'],
        'SIGNIFICANT_LATE_PAYMENT': ['mean'],
        'LATE_PAYMENT_RATIO': ['mean'],
        #'DPD_7': ['mean'],
        #'DPD_15': ['mean'],
        }

    INSTALLMENTS_60_AGG = {
        'SK_ID_PREV': ['size'],
        'DAYS_ENTRY_PAYMENT': ['max', 'mean'],
        'AMT_INSTALMENT': ['min', 'max', 'mean', 'sum'],
        'AMT_PAYMENT': ['min', 'max', 'mean', 'sum'],
        #'DPD': ['var'],
        'DBD': ['max', 'mean', 'var'],
        'PAYMENT_DIFFERENCE': ['mean'],
        #'PAYMENT_RATIO': ['mean'],
        'LATE_PAYMENT': ['mean'],
        'SIGNIFICANT_LATE_PAYMENT': ['mean'],
        'LATE_PAYMENT_RATIO': ['mean'],
        #'DPD_7': ['mean'],
        #'DPD_15': ['mean'],
        }
    
    # Installments in the last x months
    pay_36m_agg = group(df[df['DAYS_INSTALMENT'] >= -30*36], 'INS_36M_',   #agg df
                     INSTALLMENTS_36_AGG)
    pay_60m_agg = group(df[df['DAYS_INSTALMENT'] >= -30*60], 'INS_36M_',   #agg df
                     INSTALLMENTS_60_AGG)
    
    # Last x periods trend features
    group_features = ['SK_ID_PREV', 'DPD', 'LATE_PAYMENT',
                      'PAID_OVER_AMOUNT', 'PAID_OVER', 'DAYS_INSTALMENT']
    gp = df[group_features].reset_index()
    # INSTALLMENTS TREND PERIODS
    #INSTALLMENTS_LAST_K_TREND_PERIODS =  [12, 24, 60, 120]
    #func = partial(trend_in_last_k_instalment_features, periods= INSTALLMENTS_LAST_K_TREND_PERIODS)
    #g = parallel_apply(gp, func, index_name='SK_ID_CURR', chunk_size=10000).reset_index()
    #pay_agg = pay_agg.merge(g, on='SK_ID_CURR', how='left')

    # Last loan features
    INSTALLMENTS_AGG2 = {
        'DPD': ['sum', 'mean', 'max', 'std'],
        'LATE_PAYMENT': ['count', 'mean'],
        'PAID_OVER_AMOUNT': ['sum', 'mean', 'max', 'min', 'std'],
        'PAID_OVER': ['count'],
        }
    last_id = gp[['SK_ID_CURR', 'DAYS_INSTALMENT']].groupby('SK_ID_CURR').max().merge(
        gp[['SK_ID_CURR','DAYS_INSTALMENT','SK_ID_PREV']], on=['SK_ID_CURR', 'DAYS_INSTALMENT'])['SK_ID_PREV'].values

    last_agg = group(gp[gp['SK_ID_PREV'].isin(last_id)], 'LAST_LOAN_', INSTALLMENTS_AGG2)  #agg df
    #import pdb; pdb.set_trace()
    #pay_agg = pay_agg.merge(g, on='SK_ID_CURR', how='left')
    agg_df = reduce(lambda left, right: pd.merge(left, right, on='SK_ID_CURR', how='left'),
                    (pay_agg, pay_36m_agg, pay_60m_agg, last_agg))
    del pay_agg, pay_36m_agg, pay_60m_agg, last_agg, gp
    return agg_df.set_index('SK_ID_CURR')
```

```python
def get_prev(path, nrows=None):
    """preprocesses previous_application data
    Args:
        path(str): path to directory where data is in
        nrows(int): number of rows to read in (used for testing purposes)
    Returns:
        df(pd.DataFrame): dataframe of preproccessed previous_application
                          data with features engineered
    """
    df = pd.read_csv(os.path.join(path, 'previous_application.csv.zip'), nrows=nrows)
    ip = pd.read_csv(os.path.join(path, 'installments_payments.csv.zip'), nrows=nrows)

    # join installments_payments SK_ID_PREV and AMT_PAYMENT features to previous applications
    df = df.merge(ip[['SK_ID_PREV', 'AMT_INSTALMENT', 'AMT_PAYMENT']].groupby('SK_ID_PREV').sum(), how='left', on='SK_ID_PREV')

    # get SK_ID_PREV for loans with late payments
    ip['DPD'] = (ip['DAYS_ENTRY_PAYMENT'] - ip['DAYS_INSTALMENT']).clip(lower=0)
    dpd_id = ip[ip['DPD']>0]['SK_ID_PREV'].unique()
    del ip

    # basic feature eng -------------------------
    df['ACTIVE'] = (df.DAYS_TERMINATION==365243).astype('int8')
    df['INTEREST'] = df['CNT_PAYMENT']*df['AMT_ANNUITY'] - df['AMT_CREDIT']
    df['INTEREST_RATE'] = 2*12*df['INTEREST'] / (df['AMT_CREDIT']*(df['CNT_PAYMENT']+1))
    df['INTEREST_RATIO'] = df['INTEREST'] / df['AMT_CREDIT']
    df['APP_CREDIT_RATIO'] = df['AMT_APPLICATION'] / df['AMT_CREDIT']
    df['APP_CREDIT_DIFF'] = df['AMT_APPLICATION'] - df['AMT_CREDIT']
    df['APP_CREDIT_ANNUITY_RATIO'] = df['AMT_CREDIT'] / df['AMT_ANNUITY']
    df['DOWN_PAYMENT_CREDIT_RATIO'] = df['AMT_DOWN_PAYMENT'] / df['AMT_CREDIT']

    # active previous app ------------------------
    prev_active = df[(df['NAME_CONTRACT_STATUS']=='Approved')&(df['ACTIVE']==1)]

    #impute with zero to calculate DEBT
    prev_active.fillna({'AMT_PAYMENT':0, 'AMT_DOWN_PAYMENT':0}, inplace=True)
    
    prev_active["LOAN_BALANCE"] = (prev_active.AMT_CREDIT - (prev_active.AMT_DOWN_PAYMENT + prev_active.AMT_PAYMENT)).clip(lower=0)
    prev_active['REPAYMENT_RATIO'] = prev_active['AMT_PAYMENT'] / prev_active['AMT_CREDIT']

    PREVIOUS_ACTIVE_AGG = {
    'SK_ID_PREV': ['nunique'],
    'AMT_ANNUITY': ['sum', 'max', 'min', 'mean', 'median'],
    'AMT_APPLICATION': ['sum', 'max', 'min', 'mean', 'median'],
    'AMT_CREDIT': ['sum', 'max', 'min', 'mean', 'median'],
    #'AMT_DOWN_PAYMENT': ['max', 'mean'],
    'DAYS_DECISION': ['min', 'mean'],
    'CNT_PAYMENT': ['mean', 'max', 'sum'],
    'DAYS_LAST_DUE_1ST_VERSION': ['min', 'max', 'mean'],
    'AMT_PAYMENT': ['sum', 'max', 'min', 'mean'],
    'APP_CREDIT_RATIO': ['min', 'mean'],
    'INTEREST': ['sum', 'mean'],
    'INTEREST_RATE': ['max', 'min', 'mean'],
    'INTEREST_RATIO': ['max', 'min', 'mean'],
    'LOAN_BALANCE' : ['max', 'mean'],
    'REPAYMENT_RATIO': ['mean'],    
    }

    prev_active_agg = group(prev_active, 'PREV_ACTIVE_', PREVIOUS_ACTIVE_AGG) #agg df
    del prev_active

    #cleaning
    df.set_index('SK_ID_CURR', inplace=True)
    df.replace(365243, float('inf'), inplace=True)
    df['DAYS_LAST_DUE_DIFF'] = df['DAYS_LAST_DUE_1ST_VERSION'] - df['DAYS_LAST_DUE']
    
    #other------------
    PREVIOUS_AGG = {
        'SK_ID_PREV': ['nunique'],
        'AMT_ANNUITY': ['max', 'mean'],
        'AMT_DOWN_PAYMENT': ['max'],
        'HOUR_APPR_PROCESS_START': ['max', 'mean'],
        'RATE_DOWN_PAYMENT': ['max', 'mean'],
        'DAYS_DECISION': ['min', 'max', 'mean'],
        'CNT_PAYMENT': ['max', 'mean'],
        #'DAYS_TERMINATION': ['max'],
        # Engineered features
        'APP_CREDIT_ANNUITY_RATIO': ['mean', 'max'],
        'APP_CREDIT_DIFF': ['min', 'max', 'mean'],
        'APP_CREDIT_RATIO': ['min', 'max', 'mean'],
        'DOWN_PAYMENT_CREDIT_RATIO': ['mean']
    }
    #import pdb; pdb.set_trace()
    cat_cols = df.dtypes[df.dtypes=='O'].index.values
    df = pd.get_dummies(df, dummy_na=True, columns=cat_cols)
    cat_cols = df.dtypes[df.dtypes=='O'].index.values
    CAT_AGG = {key: ['mean', 'sum'] for key in cat_cols}
    prev_agg = group(df, 'PREV_', {**PREVIOUS_AGG, **CAT_AGG}) #agg df

    # aggregeations for approved and refused loans------
    PREV_APPROVED_AGG = {
        'SK_ID_PREV': ['nunique'],
        'AMT_ANNUITY': ['min', 'max', 'mean'],
        'AMT_CREDIT': ['min', 'max', 'mean'],
        'AMT_DOWN_PAYMENT': ['max'],
        'AMT_GOODS_PRICE': ['max'],
        'HOUR_APPR_PROCESS_START': ['min', 'max'],
        'DAYS_DECISION': ['mean'],
        'CNT_PAYMENT': ['max', 'mean'],
        #'DAYS_TERMINATION': ['mean'],
        # Engineered features
        'APP_CREDIT_ANNUITY_RATIO': ['mean', 'max'],
        'APP_CREDIT_DIFF': ['max'],
        'APP_CREDIT_RATIO': ['min', 'max'],
        # The following features are only for approved applications
        #'DAYS_FIRST_DRAWING': ['max', 'mean'],
        'DAYS_FIRST_DUE': ['min'],
        'DAYS_LAST_DUE_1ST_VERSION': ['min', 'mean'],
        #'DAYS_LAST_DUE': ['max', 'mean'],
        #'DAYS_LAST_DUE_DIFF': ['min', 'max', 'mean'],
        'INTEREST': ['max', 'mean'],
    }

    PREV_REFUSED_AGG = {
        'AMT_APPLICATION': ['max'],
        'AMT_CREDIT': ['min', 'max'],
        'DAYS_DECISION': ['max', 'mean'],
        'CNT_PAYMENT': ['mean'],
        # Engineered features
        'APP_CREDIT_DIFF': ['min', 'max', 'mean', 'var'],
        'APP_CREDIT_RATIO': ['min', 'mean'],
        'NAME_CONTRACT_TYPE_Consumer loans': ['mean'],
        'NAME_CONTRACT_TYPE_Cash loans': ['mean'],
        'NAME_CONTRACT_TYPE_Revolving loans': ['mean'],
    }
    approved = df[df['NAME_CONTRACT_STATUS_Approved']==1]
    refused = df[df['NAME_CONTRACT_STATUS_Refused']==1]
    prev_approved_agg = group(approved, 'PREV_APPROVED_', PREV_APPROVED_AGG) #agg df
    del approved
    prev_refused_agg = group(refused, 'PREV_REFUSED_', PREV_REFUSED_AGG) #agg df
    del refused

    # aggregations for consumer loans and cash loans-----------------
    PREVIOUS_CONSUMER_AGG = {
        'AMT_CREDIT': ['sum'],
        'AMT_ANNUITY': ['mean', 'max'],
        'INTEREST': ['mean', 'var'],
        'APP_CREDIT_DIFF': ['var'],
        'APP_CREDIT_RATIO': ['max', 'mean'],
        'DAYS_DECISION': ['max'],
        'DAYS_LAST_DUE_1ST_VERSION': ['max', 'mean'],
        'CNT_PAYMENT': ['mean'],
    }

    PREVIOUS_CASH_AGG = {
        'AMT_CREDIT': ['sum'],
        'AMT_ANNUITY': ['mean'],
        'INTEREST': ['min', 'mean', 'max', 'var'],
        'APP_CREDIT_DIFF': ['min', 'var'],
        'APP_CREDIT_RATIO': ['min', 'max'],
        #'DAYS_DECISION': ['max'],
        'DAYS_LAST_DUE_1ST_VERSION': ['max', 'mean'],
        'CNT_PAYMENT': ['mean'],
    }

    prev_consumer_agg = group(df[df['NAME_CONTRACT_TYPE_Consumer loans']==1], #agg df
                              'PREV_CONSUMER_', PREVIOUS_CONSUMER_AGG)
    prev_cash_agg = group(df[df['NAME_CONTRACT_TYPE_Cash loans']==1],         #agg df
                          'PREV_CASH_', PREVIOUS_CASH_AGG)
    
    # aggregations for loans with late payments---------------------
    PREV_LATE_PAYMENTS_AGG = {
        'DAYS_DECISION': ['min', 'max', 'mean'],
        #'DAYS_LAST_DUE_1ST_VERSION': ['min', 'max', 'mean'],
        # Engineered features
        'APP_CREDIT_DIFF': ['min'],
        'NAME_CONTRACT_TYPE_Consumer loans': ['mean'],
        'NAME_CONTRACT_TYPE_Cash loans': ['mean'],
        'NAME_CONTRACT_TYPE_Revolving loans': ['mean'],
    }
    prev_late_agg = group(df[df['SK_ID_PREV'].isin(dpd_id)],    #agg df
                          'PREV_LATE_', PREV_LATE_PAYMENTS_AGG)
    
    # aggregations for loans in the last x months---------------------
    PREVIOUS_LAST12M_AGG = {
        'AMT_CREDIT': ['sum'],
        'AMT_ANNUITY': ['mean', 'max'],
        'INTEREST': ['mean', 'max'],
        'DAYS_DECISION': ['mean'],
        'DAYS_LAST_DUE_1ST_VERSION': ['min'],
        # Engineered features
        #'APP_CREDIT_DIFF': ['min'],
        'APP_CREDIT_RATIO': ['min'],
        'NAME_CONTRACT_TYPE_Consumer loans': ['mean'],
        'NAME_CONTRACT_TYPE_Cash loans': ['mean'],
        #'NAME_CONTRACT_TYPE_Revolving loans': ['mean'],
    }
    prev_last12m_agg = group(df[df['DAYS_DECISION'] >= -30*12], 'PREV_LAST12M_', #agg df
                             PREVIOUS_LAST12M_AGG)
    
    PREVIOUS_LAST24M_AGG = {
        'AMT_CREDIT': ['sum'],
        'AMT_ANNUITY': ['mean', 'max'],
        'INTEREST': ['mean', 'max'],
        'DAYS_DECISION': ['min', 'mean'],
        'DAYS_LAST_DUE_1ST_VERSION': ['min'],
        # Engineered features
        'APP_CREDIT_DIFF': ['min'],
        'APP_CREDIT_RATIO': ['min', 'max', 'mean'],
        'NAME_CONTRACT_TYPE_Consumer loans': ['mean'],
        #'NAME_CONTRACT_TYPE_Cash loans': ['mean'],
        #'NAME_CONTRACT_TYPE_Revolving loans': ['mean'],
    }
    prev_last24m_agg = group(df[df['DAYS_DECISION'] >= -30*24], 'PREV_LAST24M_', #agg df
                             PREVIOUS_LAST24M_AGG)

    agg_df = reduce(lambda left, right: pd.merge(left, right, on='SK_ID_CURR', how='left'),
                    (prev_agg, prev_active_agg, prev_approved_agg, prev_refused_agg, 
                     prev_consumer_agg, prev_cash_agg, prev_late_agg, 
                     prev_last12m_agg, prev_last24m_agg))
    
    #clean up
    del df, prev_agg, prev_active_agg, prev_approved_agg, prev_refused_agg
    del prev_consumer_agg, prev_cash_agg, prev_late_agg, prev_last12m_agg, prev_last24m_agg

    #agg_df['SK_ID_CURR'] = agg_df['SK_ID_CURR'].astype('int')

    return agg_df.set_index('SK_ID_CURR')
```

```python
def get_bureau_balance(path, nrows=None):
    """preprocesses bureau_balance data
    Args:
        path(str): path to directory where data is in
        nrows(int): number of rows to read in (used for testing purposes)
    Returns:
        df(pd.DataFrame): dataframe of preproccessed bureau_balance
                          data with features engineered
    """
    df = pd.read_csv(os.path.join(path, 'bureau_balance.csv.zip'), nrows=nrows)

    # get max status features ------------
    df2 = df.replace({'C': 0, 'X': 0})
    df2.STATUS = df2.STATUS.astype('int8')
    df2.rename(columns={'STATUS': 'STATUS_MAX', 'MONTHS_BALANCE': 'MONTHS_BALANCE_LAST'}, inplace=True)
    df2 = df2.groupby('SK_ID_BUREAU').max()

    # get one hot encode features ---------------
    BB_AGG = {'MONTHS_BALANCE': ['min', 'max', 'mean', 'size']}
    df = pd.get_dummies(df)
    cat_cols = df.dtypes[df.dtypes=='O'].values
    df3 = df.groupby('SK_ID_BUREAU').mean().reset_index()
    df = group(df, '', BB_AGG, 'SK_ID_BUREAU')
    
    df_agg = reduce(lambda left, right: pd.merge(left, right, how='left', on='SK_ID_BUREAU'),
                  (df, df2, df3))
    del df, df2, df3
    return df_agg
```

## bureau



```python
def get_bureau(path, nrows=None):
    """preprocesses bureau data
    Args:
        path(str): path to directory where data is in
        nrows(int): number of rows to read in (used for testing purposes)
    Returns:
        df(pd.DataFrame): dataframe of preproccessed bureau
                          data with features engineered
    """
    df = pd.read_csv(os.path.join(path, 'bureau.csv.zip'), nrows=nrows)
    bb = get_bureau_balance(path, nrows)
    df = df.merge(bb, how='left', on='SK_ID_BUREAU')

    # Credit duration and credit/account end date difference
    df['CREDIT_DURATION'] = -df['DAYS_CREDIT'] + df['DAYS_CREDIT_ENDDATE']
    df['ENDDATE_DIF'] = df['DAYS_CREDIT_ENDDATE'] - df['DAYS_ENDDATE_FACT']
    # Credit to debt ratio and difference
    df['DEBT_PERCENTAGE'] = df['AMT_CREDIT_SUM'] / df['AMT_CREDIT_SUM_DEBT']
    df['DEBT_CREDIT_DIFF'] = df['AMT_CREDIT_SUM'] - df['AMT_CREDIT_SUM_DEBT']
    df['CREDIT_TO_ANNUITY_RATIO'] = df['AMT_CREDIT_SUM'] / df['AMT_ANNUITY']

    # one-hot-encode
    df = pd.get_dummies(df)

    #import pdb; pdb.set_trace()
    # Flag months with late payments (days past due)
    df['STATUS_12345'] = 0
    for i in range(1,6):
        df['STATUS_12345'] += df['STATUS_{}'.format(i)]

    # Aggregate by number of months in balance and merge with bureau (loan length agg)
    features = ['AMT_CREDIT_MAX_OVERDUE', 'AMT_CREDIT_SUM_OVERDUE', 'AMT_CREDIT_SUM',
        'AMT_CREDIT_SUM_DEBT', 'DEBT_PERCENTAGE', 'DEBT_CREDIT_DIFF', 'STATUS_0', 'STATUS_12345']
    agg_length = df.groupby('MONTHS_BALANCE_SIZE')[features].mean().reset_index()
    agg_length.rename(columns={feat: 'LL_' + feat for feat in features}, inplace=True)
    df = df.merge(agg_length, how='left', on='MONTHS_BALANCE_SIZE')
    del agg_length

    # General loans aggregations ------------------------
    BUREAU_AGG = {
        'SK_ID_BUREAU': ['nunique'],
        'DAYS_CREDIT': ['min', 'mean'],
        'DAYS_CREDIT_ENDDATE': ['max'],
        'AMT_CREDIT_MAX_OVERDUE': ['max', 'mean'],
        'AMT_CREDIT_SUM': ['max', 'mean', 'sum'],
        'AMT_CREDIT_SUM_DEBT': ['mean', 'sum'],
        'AMT_CREDIT_SUM_OVERDUE': ['mean', 'sum'],
        'AMT_ANNUITY': ['mean'],
        'DEBT_CREDIT_DIFF': ['mean'],
        'MONTHS_BALANCE_MEAN': ['mean', 'var'],
        'MONTHS_BALANCE_SIZE': ['mean', 'sum'],
        # Categorical
        'STATUS_0': ['sum', 'mean'],
        'STATUS_1': ['sum', 'mean'],
        'STATUS_12345': ['sum', 'mean'],
        'STATUS_C': ['sum', 'mean'],
        'STATUS_X': ['sum', 'mean'],
        'CREDIT_ACTIVE_Active': ['sum'],
        'CREDIT_ACTIVE_Closed': ['sum'],
        'CREDIT_ACTIVE_Sold': ['sum', 'mean'],
        'CREDIT_TYPE_Consumer credit': ['sum'],
        'CREDIT_TYPE_Credit card': ['sum'],
        'CREDIT_TYPE_Car loan': ['sum', 'mean'],
        'CREDIT_TYPE_Mortgage': ['sum'],
        'CREDIT_TYPE_Microloan': ['sum'],
        # Group by loan duration features (months)
        #'LL_AMT_CREDIT_SUM_OVERDUE': ['mean'],
        'LL_DEBT_CREDIT_DIFF': ['mean'],
        'LL_STATUS_12345': ['mean'],
    }
    bur_agg = group(df, 'BUR_', BUREAU_AGG)                                   # agg df

    # active loans aggregations ------------------------
    BUREAU_ACTIVE_AGG = {
        #'DAYS_CREDIT': ['max', 'mean'],
        'DAYS_CREDIT_ENDDATE': ['min', 'max'],
        'AMT_CREDIT_MAX_OVERDUE': ['max'],
        'AMT_CREDIT_SUM': ['mean', 'min', 'sum'],
        'AMT_CREDIT_SUM_DEBT': ['mean', 'sum'],
        'AMT_CREDIT_SUM_OVERDUE': ['mean', 'sum'],
        'AMT_CREDIT_SUM_LIMIT': ['sum', 'mean', 'max'], 
        'DAYS_CREDIT_UPDATE': ['mean'],
        'DEBT_PERCENTAGE': ['mean'],
        'DEBT_CREDIT_DIFF': ['mean'],
        #'CREDIT_TO_ANNUITY_RATIO': ['mean'],
        'MONTHS_BALANCE_MEAN': ['mean'],
        #'MONTHS_BALANCE_SIZE': ['mean', 'sum'],
        'CNT_CREDIT_PROLONG': ['sum']
    }
    bur_active_agg = group(df[df['CREDIT_ACTIVE_Active'] == 1], 'BUR_ACTIVE_',  #agg df 
                           BUREAU_ACTIVE_AGG)

    # closed loans aggregations -----------------------
    BUREAU_CLOSED_AGG = {
        'DAYS_CREDIT': ['max', 'var'],
        'DAYS_CREDIT_ENDDATE': ['max'],
        'AMT_CREDIT_MAX_OVERDUE': ['max', 'mean'],
        #'AMT_CREDIT_SUM_OVERDUE': ['mean'],
        'AMT_CREDIT_SUM': ['max', 'sum'],
        'AMT_CREDIT_SUM_DEBT': ['max'],
        #'DAYS_CREDIT_UPDATE': ['max'],
        'ENDDATE_DIF': ['mean'],
        'STATUS_12345': ['mean'],
    }
    bur_closed_agg = group(df[df['CREDIT_ACTIVE_Closed'] == 1], 'BUR_CLOSED_',   #agg df
                           BUREAU_CLOSED_AGG)

    # loan types aggregations----------------------------
    BUREAU_CONSUMER_AGG = {
        'DAYS_CREDIT': ['mean', 'max'],
        'AMT_CREDIT_MAX_OVERDUE': ['mean'],
        'AMT_CREDIT_SUM': ['mean', 'max'],
        'AMT_CREDIT_SUM_DEBT': ['mean'],
        #'DEBT_PERCENTAGE': ['mean'],
        'DEBT_CREDIT_DIFF': ['mean'],
        'DAYS_CREDIT_ENDDATE': ['max'],
    }
    bur_consumer_agg = group(df[df['CREDIT_TYPE_Consumer credit']==1], 'BUR_CONSUMER_', #agg df
                             BUREAU_CONSUMER_AGG)
    
    BUREAU_CREDIT_AGG = {
        'DAYS_CREDIT': ['max'],
        'AMT_CREDIT_MAX_OVERDUE': ['max'],
        'AMT_CREDIT_SUM': ['max'],
        'AMT_CREDIT_SUM_DEBT': ['max'],
        'DEBT_PERCENTAGE': ['mean'],
        'DEBT_CREDIT_DIFF': ['mean'],
        #'DAYS_CREDIT_ENDDATE': ['max'],
    }
    bur_credit_agg = group(df[df['CREDIT_TYPE_Credit card']==1], 'BUR_CREDIT_', #agg df
                             BUREAU_CREDIT_AGG)
    
    BUREAU_MORTGAGE_AGG = {
        'DAYS_CREDIT': ['mean'],
        #'AMT_CREDIT_MAX_OVERDUE': ['mean', 'max'],
        'AMT_CREDIT_SUM': ['mean'],
        'AMT_CREDIT_SUM_DEBT': ['max'],
        'DEBT_PERCENTAGE': ['mean'],
        'DEBT_CREDIT_DIFF': ['mean'],
        'DAYS_CREDIT_ENDDATE': ['max'],
    }
    bur_mortgage_agg = group(df[df['CREDIT_TYPE_Mortgage']==1], 'BUR_MORTGAGE_', #agg df
                             BUREAU_MORTGAGE_AGG)
    
    BUREAU_CAR_AGG = {
        'DAYS_CREDIT': ['mean', 'max'],
        'AMT_CREDIT_MAX_OVERDUE': ['mean'],
        #'AMT_CREDIT_SUM': ['mean', 'max'],
        'AMT_CREDIT_SUM_DEBT': ['mean'],
        'DEBT_PERCENTAGE': ['mean'],
        'DEBT_CREDIT_DIFF': ['mean'],
        'DAYS_CREDIT_ENDDATE': ['max'],
    }
    bur_car_agg = group(df[df['CREDIT_TYPE_Car loan']==1], 'BUR_CAR_',          #agg df
                             BUREAU_CAR_AGG)
    
    BUREAU_MICROLOAN_AGG = {
        'DAYS_CREDIT': ['mean'],
        #'AMT_CREDIT_MAX_OVERDUE': ['mean', 'max'],
        'AMT_CREDIT_SUM': ['mean', 'max'],
        'AMT_CREDIT_SUM_DEBT': ['mean'],
        'DEBT_PERCENTAGE': ['mean'],
        'DEBT_CREDIT_DIFF': ['mean'],
        #'DAYS_CREDIT_ENDDATE': ['max'],
    }
    bur_microloan_agg = group(df[df['CREDIT_TYPE_Microloan']==1], 'BUR_MICROLOAN_', #agg df
                             BUREAU_MICROLOAN_AGG)
    
    # time based aggregations: last x months
    BUREAU_LAST6M_AGG = {
        'AMT_CREDIT_MAX_OVERDUE': ['mean'],
        #'AMT_CREDIT_SUM_OVERDUE': ['mean'],
        'AMT_CREDIT_SUM': ['max', 'sum'],
        'AMT_CREDIT_SUM_DEBT': ['mean', 'sum'],
        #'DEBT_PERCENTAGE': ['mean'],
        #'DEBT_CREDIT_DIFF': ['mean'],
        'STATUS_0': ['mean'],
        #'STATUS_12345': ['mean'],
    }
    bur_last6m_agg = group(df[df['DAYS_CREDIT'] >= -30*6], 'BUR_LAST6M_',       #agg df
                           BUREAU_LAST6M_AGG)
    
    BUREAU_LAST12M_AGG = {
        'AMT_CREDIT_MAX_OVERDUE': ['mean'],
        #'AMT_CREDIT_SUM_OVERDUE': ['mean'],
        'AMT_CREDIT_SUM': ['max', 'sum'],
        'AMT_CREDIT_SUM_DEBT': ['mean', 'sum'],
        #'DEBT_PERCENTAGE': ['mean'],
        'DEBT_CREDIT_DIFF': ['mean'],
        'STATUS_0': ['mean'],
        'STATUS_12345': ['mean'],
    }
    bur_last12m_agg = group(df[df['DAYS_CREDIT'] >= -30*12], 'BUR_LAST12M_',     #agg df
                           BUREAU_LAST12M_AGG)

    # last loan max overdue----------------------
    bur_lastmax_agg = df.sort_values(['DAYS_CREDIT']).groupby('SK_ID_CURR')['AMT_CREDIT_MAX_OVERDUE'].last().reset_index()
    bur_lastmax_agg.rename(columns={'AMT_CREDIT_MAX_OVERDUE': 'BUREAU_LAST_LOAN_MAX_OVERDUE'}, inplace=True) #agg df

    agg_df = reduce(lambda left, right: pd.merge(left, right, on='SK_ID_CURR', how='left'),
                    (bur_agg, bur_active_agg, bur_closed_agg, bur_consumer_agg,
                     bur_credit_agg, bur_mortgage_agg, bur_car_agg,
                     bur_microloan_agg, bur_last6m_agg, bur_last12m_agg, 
                     bur_lastmax_agg))
    del bur_agg, bur_active_agg, bur_closed_agg, bur_consumer_agg
    del bur_credit_agg, bur_mortgage_agg, bur_car_agg, bur_microloan_agg,
    del bur_last6m_agg, bur_last12m_agg, bur_lastmax_agg

    # Ratios: total debt/total credit and active loans debt/ active loans credit
    agg_df['BUR_DEBT_OVER_CREDIT'] = agg_df['BUR_AMT_CREDIT_SUM_DEBT_SUM']/agg_df['BUR_AMT_CREDIT_SUM_SUM']
    agg_df['BUR_ACTIVE_DEBT_OVER_CREDIT'] = agg_df['BUR_ACTIVE_AMT_CREDIT_SUM_DEBT_SUM']/agg_df['BUR_ACTIVE_AMT_CREDIT_SUM_SUM']
    
    return agg_df.set_index('SK_ID_CURR')
```

## credit_card_balance

```python
def get_credit(path, nrows=None):
    """preprocesses credit_card_balance data
    Args:
        path(str): path to directory where data is in
        nrows(int): number of rows to read in (used for testing purposes)
    Returns:
        df(pd.DataFrame): dataframe of preproccessed credit_card_balance
                          data with features engineered
    """
    df = pd.read_csv(os.path.join(path, 'credit_card_balance.csv.zip'), nrows=nrows)
    df.rename(columns={'AMT_RECIVABLE': 'AMT_RECEIVABLE'}, inplace=True)
    # Amount used from limit
    df['LIMIT_USE'] = df['AMT_BALANCE'] / df['AMT_CREDIT_LIMIT_ACTUAL']
    # Current payment / Min payment
    df['PAYMENT_DIV_MIN'] = df['AMT_PAYMENT_CURRENT'] / df['AMT_INST_MIN_REGULARITY']
    # Late payment
    df['LATE_PAYMENT'] = (df['SK_DPD']>0).astype('int8')
    # How much drawing of limit
    df['DRAWING_LIMIT_RATIO'] = df['AMT_DRAWINGS_ATM_CURRENT'] / df['AMT_CREDIT_LIMIT_ACTUAL']

    #one-hot-encode
    df = pd.get_dummies(df)

    # basic aggregations ----------------------
    CREDIT_CARD_AGG = {
        'MONTHS_BALANCE': ['min'],
        'AMT_BALANCE': ['max'],
        #'AMT_CREDIT_LIMIT_ACTUAL': ['max'],
        'AMT_DRAWINGS_ATM_CURRENT': ['max', 'sum'],
        'AMT_DRAWINGS_CURRENT': ['max', 'sum'],
        'AMT_DRAWINGS_POS_CURRENT': ['max', 'sum'],
        'AMT_INST_MIN_REGULARITY': ['max', 'mean'],
        'AMT_PAYMENT_TOTAL_CURRENT': ['max', 'mean', 'sum', 'var'],
        'AMT_TOTAL_RECEIVABLE': ['max', 'mean'],
        'CNT_DRAWINGS_ATM_CURRENT': ['max', 'mean', 'sum'],
        'CNT_DRAWINGS_CURRENT': ['max', 'mean', 'sum'],
        #'CNT_DRAWINGS_POS_CURRENT': ['mean'],
        'SK_DPD': ['max'],
        'SK_DPD_DEF': ['max', 'sum'],
        'LIMIT_USE': ['max'],
        'PAYMENT_DIV_MIN': ['min', 'mean'],
        'LATE_PAYMENT': ['sum'],
    }
    cc_agg = group(df, 'CC_', CREDIT_CARD_AGG)      # agg df

    # active aggregations ------------------------------
    CREDIT_CARD_ACTIVE_AGG = {
        'MONTHS_BALANCE': ['min'],
        'AMT_BALANCE': ['max'],
        #'AMT_CREDIT_LIMIT_ACTUAL': ['max'],
        'AMT_DRAWINGS_ATM_CURRENT': ['max', 'sum'],
        'AMT_DRAWINGS_CURRENT': ['max', 'sum'],
        'AMT_DRAWINGS_POS_CURRENT': ['max', 'sum'],
        'AMT_INST_MIN_REGULARITY': ['max', 'mean'],
        'AMT_PAYMENT_TOTAL_CURRENT': ['max', 'mean', 'sum', 'var'],
        'AMT_TOTAL_RECEIVABLE': ['max', 'mean'],
        'CNT_DRAWINGS_ATM_CURRENT': ['mean', 'sum'],
        'CNT_DRAWINGS_CURRENT': ['max', 'mean', 'sum'],
        'CNT_DRAWINGS_POS_CURRENT': ['mean'],
        'SK_DPD': ['sum'],
        'SK_DPD_DEF': ['max', 'sum'],
        'LIMIT_USE': ['max'],
        'PAYMENT_DIV_MIN': ['min'],
        'LATE_PAYMENT': ['max'],
    }
    cc_active_agg = group(df[df['NAME_CONTRACT_STATUS_Active']==1], 'CC_ACTIVE_',   # agg df
                          CREDIT_CARD_ACTIVE_AGG)

    # last month balance aggregations ------------------
    CREDIT_CARD_LAST_AGG = {
        'AMT_BALANCE': ['mean', 'max', 'sum']
    }
    last_ids = df.groupby('SK_ID_PREV')['MONTHS_BALANCE'].idxmax()
    cc_last_agg = group(df[df.index.isin(last_ids)], 'CC_LAST_',      # agg df
                        CREDIT_CARD_LAST_AGG)
    
    # last x months aggregations-------------------------
    CREDIT_CARD_12M_AGG = {
        'CNT_DRAWINGS_ATM_CURRENT': ['mean'],
        'SK_DPD': ['max', 'sum'],
        'AMT_BALANCE': ['mean', 'max'],
        'LIMIT_USE': ['max', 'mean']
    }
    cc_12m_agg = group(df[df['MONTHS_BALANCE'] >= -12], 'CC_12M_',    # agg df
                       CREDIT_CARD_12M_AGG)
    
    CREDIT_CARD_24M_AGG = {
        'CNT_DRAWINGS_ATM_CURRENT': ['mean'],
        'SK_DPD': ['sum'],
        'AMT_BALANCE': ['mean', 'max'],
        'LIMIT_USE': ['max', 'mean']
    }
    cc_24m_agg = group(df[df['MONTHS_BALANCE'] >= -24], 'CC_24M_',    # agg df
                       CREDIT_CARD_24M_AGG)
    
    CREDIT_CARD_48M_AGG = {
        'CNT_DRAWINGS_ATM_CURRENT': ['mean'],
        'SK_DPD': ['max', 'sum'],
        'AMT_BALANCE': ['mean', 'max'],
        'LIMIT_USE': ['max', 'mean']
    }
    cc_48m_agg = group(df[df['MONTHS_BALANCE'] >= -48], 'CC_48M_',    # agg df
                       CREDIT_CARD_48M_AGG)
    
    df_agg = reduce(lambda left, right: pd.merge(left, right, how='left', on='SK_ID_CURR'),
                    (cc_agg, cc_active_agg, cc_last_agg, cc_12m_agg, cc_24m_agg,
                     cc_48m_agg))
    del cc_agg, cc_active_agg, cc_last_agg, cc_12m_agg, cc_24m_agg, cc_48m_agg
    return df_agg.set_index('SK_ID_CURR')
```

## POS_CASH_balance

```python
def get_pos_cash(path, nrows=None):
    """preprocesses POS_CASH_balance data
    Args:
        path(str): path to directory where data is in
        nrows(int): number of rows to read in (used for testing purposes)
    Returns:
        df(pd.DataFrame): dataframe of preproccessed POS_CASH_balance
                          data with features engineered
    """
    df = pd.read_csv(os.path.join(path, 'POS_CASH_balance.csv.zip'), nrows=nrows)
    df['LATE_PAYMENT'] = (df['SK_DPD']>0).astype('int8')
    df = get_group_stats(df, ['SK_ID_PREV'], 'LATE_PAYMENT', 'sum', 'LATE_PAYMENT_SUM').reset_index()

    # one-hot-encode
    df = pd.get_dummies(df)

    # basic aggregations -----------------------------------
    POS_CASH_AGG = {
        'SK_ID_PREV': ['nunique'],
        'MONTHS_BALANCE': ['min', 'max'],
        'SK_DPD': ['mean', 'sum'],
        'SK_DPD_DEF': ['max', 'mean', 'sum'],
        'LATE_PAYMENT': ['mean'],
        # categorical
        'NAME_CONTRACT_STATUS_Active': ['mean', 'sum'],
        'NAME_CONTRACT_STATUS_Amortized debt': ['mean', 'sum'],
        'NAME_CONTRACT_STATUS_Approved': ['sum'], 
        'NAME_CONTRACT_STATUS_Canceled': ['mean', 'sum'],
        'NAME_CONTRACT_STATUS_Completed': ['mean', 'sum'], 
        'NAME_CONTRACT_STATUS_Demand': ['mean', 'sum'],
        'NAME_CONTRACT_STATUS_Returned to the store': ['mean', 'sum'],
        'NAME_CONTRACT_STATUS_Signed': ['mean', 'sum'], 
        'NAME_CONTRACT_STATUS_XNA': ['mean', 'sum'] 
    }
    pos1_agg = group(df, 'POS_', POS_CASH_AGG)                      # agg df

    # other aggregations ---------------------------
    POS_CASH_AGG2 = {
        'POS_LOAN_COMPLETED_MEAN': ['sum'], 
        'POS_COMPLETED_BEFORE_MEAN': ['sum'],
        'POS_REMAINING_INSTALMENTS': ['sum'], 
        'POS_REMAINING_INSTALMENTS_RATIO': ['sum']
    }

    OTHER_AGG = {
        'SK_ID_CURR': ['first'],
        'NAME_CONTRACT_STATUS_Completed': ['mean'],
        'CNT_INSTALMENT': ['first', 'last'],
        'CNT_INSTALMENT_FUTURE': ['last']
    }
    sort_pos = df.sort_values(by=['SK_ID_PREV', 'MONTHS_BALANCE'])
    df2 = group(sort_pos, '', 
            OTHER_AGG, 'SK_ID_PREV').rename(columns={'SK_ID_CURR_FIRST': 'SK_ID_CURR', 
                                                     'NAME_CONTRACT_STATUS_Completed_MEAN': 'POS_LOAN_COMPLETED_MEAN',
                                                     'CNT_INSTALMENT_FUTURE_LAST': 'POS_REMAINING_INSTALMENTS'})
    df2['CNT_INSTALMENT_DIFF'] = df2['CNT_INSTALMENT_FIRST'] - df2['CNT_INSTALMENT_LAST']
    df2['POS_COMPLETED_BEFORE_MEAN'] = ((df2['CNT_INSTALMENT_DIFF'] > 0) & (df2['POS_LOAN_COMPLETED_MEAN'] > 0)).astype('uint8')
    df2['POS_REMAINING_INSTALMENTS_RATIO'] = df2['POS_REMAINING_INSTALMENTS']/df2['CNT_INSTALMENT_LAST']
    cols = ['SK_ID_CURR', 'POS_LOAN_COMPLETED_MEAN', 'POS_COMPLETED_BEFORE_MEAN',
          'POS_REMAINING_INSTALMENTS', 'POS_REMAINING_INSTALMENTS_RATIO']
    pos2_agg = group(df2[cols], 'POS_', POS_CASH_AGG2)             #agg df
    del df2

    # last aggregrations -----------------------------
    POS_LAST_AGG = {
        'MONTHS_BALANCE': ['min'],
        'SK_DPD': ['max', 'mean', 'sum', 'var'],
        'SK_DPD_DEF': ['max'],
        'LATE_PAYMENT': ['mean'],
        'LATE_PAYMENT_SUM': ['mean'],
        # categorical
        'NAME_CONTRACT_STATUS_Active': ['mean', 'sum'],
        'NAME_CONTRACT_STATUS_Amortized debt': ['mean', 'sum'],
        'NAME_CONTRACT_STATUS_Approved': ['mean', 'sum'], 
        'NAME_CONTRACT_STATUS_Canceled': ['mean', 'sum'],
        'NAME_CONTRACT_STATUS_Completed': ['mean'], 
        'NAME_CONTRACT_STATUS_Demand': ['mean', 'sum'],
        'NAME_CONTRACT_STATUS_Returned to the store': ['mean'],
        'NAME_CONTRACT_STATUS_Signed': ['sum'], 
        'NAME_CONTRACT_STATUS_XNA': ['sum'] 
    }
    # Percentage of late payments for the 3 most recent applications
    # Last month of each application
    last_month_df = df.groupby('SK_ID_PREV')['MONTHS_BALANCE'].idxmax()
    # Most recent applications (last 3)
    gp = sort_pos.iloc[last_month_df].groupby('SK_ID_CURR').tail(3)
    pos3_agg = group(gp, 'POS_LAST_', POS_LAST_AGG)             #agg df
    del last_month_df, sort_pos, gp, df

    df_agg = reduce(lambda left, right: pd.merge(left, right, how='left',
                                                 on='SK_ID_CURR'),
                    (pos1_agg, pos2_agg, pos3_agg))
    del pos1_agg, pos2_agg, pos3_agg
    return df_agg.set_index('SK_ID_CURR')
```

# ETL Process

The final ETL processing pipeline that gathers all the features engineered from each data source and combines them before the initial training of the XGBoost model.

```python
def preprocess(path, test=False):
    """preprocess ALL datasets and merges/joins them together
    Args:
        path(str): path to directory where data is in
        test(boolean): if True will read in test data instead of train
    Returns:
        df(pd.DataFrame): dataframe of preproccessed and merged/joined data
    """
    app = get_app(path, test=test) #has runtimewarning
    bur = get_bureau(path)
    cc = get_credit(path)
    ip = get_ip(path)
    prev = get_prev(path) #has SettingWithCopyWarning
    pos = get_pos_cash(path) 
    df = reduce(lambda left, right: pd.merge(left, right, how='left', on='SK_ID_CURR'),
                (app, bur, cc, ip, prev, pos))
    del app, bur, cc, ip, prev, pos
    return df
```

# Bayesian Hyperparameter Optimization

Bayesian hyperparameter optimization is used because of how much more efficient it is than grid or random search. The following are some notes left when exploring this optimization technique. At the end of this section we will have our optimized XGBoost model along with its parameters and the AUC score it produces.

https://towardsdatascience.com/a-conceptual-explanation-of-bayesian-model-based-hyperparameter-optimization-for-machine-learning-b8172278050f

"The one-sentence summary of Bayesian hyperparameter optimization is: build a probability model of the objective function and use it to select the most promising hyperparameters to evaluate in the true objective function."

* keep track of past evaluation results which they use to form a probabilistic model mapping of hyperparameters to a probability of a score on the objective function
* P(score | hyperparemters)
* algorithm
  1. Build a surrogate probability model of the objective function
  2. Find the hyperparameters that perform best on the surrogate
  3. Apply these hyperparameters to the true objective function
  4. Update the surrogate model incorporating the new results
  5.  Repeat steps 2-4 until max iterations or time is reached
* bayesian methods operate in much the same way we do: we form an initial view of he world (called a prior) and then we update our model based on new experiences (the updated model is called a posterior).


[Notes on Parameter Tuning XGBoost](https://xgboost.readthedocs.io/en/latest/tutorials/param_tuning.html)

```python
# XGB parameters
xgb_reg_params = {
    'learning_rate':    hp.choice('learning_rate',    np.arange(0.05, 0.31, 0.05)),
    'max_depth':        hp.choice('max_depth',        np.arange(2, 16, 1, dtype=int)),
    'gamma':            hp.uniform('gamma', 0, 10),
    'min_child_weight': hp.choice('min_child_weight', np.arange(1, 8, 1, dtype=int)),
    'colsample_bytree': hp.choice('colsample_bytree', np.arange(0.3, 0.8, 0.1)),
    'subsample':        hp.uniform('subsample', 0.8, 1),
    'scale_pos_weight': hp.choice('scale_pos_weight', np.arange(0.8, 12, 0.1)),
}
xgb_fit_params = {
    'eval_metric': 'auc',
    'early_stopping_rounds': 5,
    'verbose': False
}
xgb_para = dict()
xgb_para['reg_params'] = xgb_reg_params
xgb_para['fit_params'] = xgb_fit_params
xgb_para['loss_func' ] = lambda y, pred_proba: roc_auc_score(y, pred_proba) * -1
```

```python
class HPOpt(object):

    def __init__(self, df):
        #train = preprocess(df)
        X, y = df.drop(['TARGET'], axis=1), df.TARGET
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(X, y, test_size=0.15, random_state=27)

    def process(self, fn_name, space, trials, algo, max_evals):
        fn = getattr(self, fn_name)
        try:
            result = fmin(fn=fn, space=space, algo=algo, max_evals=max_evals, trials=trials)
        except Exception as e:
            return {'status': STATUS_FAIL,
                    'exception': str(e)}
        return result, trials

    def xgb_clf(self, para):
        clf = xgb.XGBClassifier(tree_method='gpu_hist', n_estimators=1000 ,seed=27, 
                              use_label_encoder=False, **para['reg_params'])
        return self.train_clf(clf, para)

    def train_clf(self, clf, para):
        clf.fit(self.X_train, self.y_train,
                eval_set=[(self.X_train, self.y_train), (self.X_test, self.y_test)],
                **para['fit_params'])
        pred_proba = clf.predict_proba(self.X_test)[:,1]
        loss = para['loss_func'](self.y_test, pred_proba)
        return {'loss': loss, 'status': STATUS_OK}
```
